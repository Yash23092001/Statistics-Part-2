Q1. What is hypothesis testing in statistics?
  -
Hypothesis testing in statistics is a formal method used to evaluate claims or assumptions about a population parameter based on sample data. It involves setting up two competing hypotheses: the null hypothesis (H₀), which represents a default position (e.g., no effect or no difference), and the alternative hypothesis (H₁), which represents the researcher's claim or the effect being tested. A test statistic is then calculated from the sample data to determine how compatible the data is with H₀. The p-value, a key component, quantifies the probability of observing the test statistic (or something more extreme) if H₀ is true. If this p-value is below a predetermined significance level (α) (commonly 0.05), H₀ is rejected in favor of H₁, suggesting the results are statistically significant. Otherwise, H₀ is not rejected, meaning there isn't enough evidence to support the alternative claim. Hypothesis testing helps researchers make data-driven decisions while accounting for uncertainty and random variation.

Q2.What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - The null hypothesis (H₀) is a default statement in statistical testing that assumes no effect, no difference, or no relationship between variables—essentially representing the status quo or a position of skepticism. For example, if testing a new drug's effectiveness, H₀ might state that the drug has no impact compared to a placebo. In contrast, the alternative hypothesis (H₁) challenges this assumption by proposing that there is an effect, difference, or relationship (e.g., the drug works better than the placebo). The key difference lies in their roles: H₀ is what researchers aim to test against (and potentially reject), while H₁ embodies the claim they seek to support. Statistical analysis evaluates whether the observed data provides sufficient evidence to reject H₀ in favor of H₁, based on probabilities (like the p-value) rather than certainties. This framework ensures conclusions are drawn rigorously, minimizing bias.

Q3.What is the significance level in hypothesis testing, and why is it important?
  - The significance level (α) in hypothesis testing is a predetermined threshold (commonly set at 0.05 or 5%) that defines the maximum probability of rejecting the null hypothesis (H₀) when it is actually true—a mistake known as a Type I error (false positive). It serves as a critical benchmark for determining whether observed data provides sufficient evidence against H₀. If the calculated p-value is less than α (e.g., p < 0.05), the result is deemed "statistically significant," leading researchers to reject H₀ in favor of the alternative hypothesis (H₁). The significance level is important because it quantifies the risk of drawing incorrect conclusions, ensuring that findings are not attributed to random chance alone. By controlling α, researchers balance the trade-off between sensitivity (detecting true effects) and reliability (avoiding false claims), maintaining rigor in statistical inference. However, a lower α (e.g., 0.01) reduces Type I errors but raises the risk of Type II errors (false negatives), where true effects are overlooked. Thus, choosing an appropriate α depends on the context and consequences of errors in a study.

Q4.What does a P-value represent in hypothesis testing?
  - In hypothesis testing, the p-value is a crucial metric that helps determine the strength of the evidence against the null hypothesis (H₀). It represents the probability of obtaining the observed sample results—or more extreme ones—if the null hypothesis were true. A small p-value (typically ≤ 0.05) suggests that the observed data is unlikely under H₀, leading researchers to reject H₀ in favor of the alternative hypothesis (H₁). Conversely, a large p-value indicates weak evidence against H₀, meaning the null hypothesis cannot be rejected. Importantly, the p-value does not quantify the probability that H₀ is true or false, nor does it measure the size or importance of an effect. Instead, it serves as a tool for deciding whether the results are statistically significant, based on a predefined significance level (α). Misinterpretations of p-values (e.g., equating them with effect magnitude or real-world relevance) are common pitfalls in statistical analysis

Q5. How do you interpret the P-value in hypothesis testing
  - The p-value in hypothesis testing is a measure of the evidence against the null hypothesis (H₀). Here’s how to interpret it:

Definition: The p-value is the probability of observing your data (or more extreme results) if the null hypothesis were true.

Key Guideline:

If p ≤ α (e.g., 0.05), reject H₀. The result is statistically significant—the data is unlikely under H₀.

If p > α, fail to reject H₀. The data does not provide strong enough evidence against H₀.

Important Nuances:
Not the Probability of H₀: The p-value does not tell you the chance H₀ is true or false; it only assesses consistency with H₀.

Effect Size ≠ Significance: A tiny p-value doesn’t mean the effect is large or important—just that it’s unlikely due to random chance.

Context Matters: Always pair p-values with confidence intervals and practical significance.

Q7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
  - In hypothesis testing, the choice between a one-tailed and two-tailed test depends on the directionality of the research question. A two-tailed test evaluates whether a parameter differs from a hypothesized value in either direction (e.g., greater or smaller), making it appropriate for non-directional hypotheses. For example, testing whether a new drug has any effect (positive or negative) on blood pressure uses a two-tailed test, with rejection regions in both tails of the distribution. In contrast, a one-tailed test examines an effect in only one specified direction (e.g., strictly greater or strictly smaller), focusing the entire rejection region in one tail. For instance, if researchers hypothesize that a drug increases recovery rates (not just changes them), a one-tailed test is used. The key distinction lies in sensitivity: one-tailed tests have greater power to detect an effect in the predicted direction but ignore effects in the opposite direction, while two-tailed tests are more conservative and detect any deviation from H₀, regardless of direction. Choosing between them hinges on prior evidence and theoretical justification—misapplication (e.g., using a one-tailed test without justification) risks overlooking meaningful effects or inflating false positives.

Q8.What is the Z-test, and when is it used in hypothesis testing?
  - The Z-test is a statistical hypothesis test used to determine whether there is a significant difference between a sample mean and a known population mean (or between the means of two samples) when the population standard deviation is known and the sample size is sufficiently large (typically n ≥ 30). It relies on the Z-distribution (standard normal distribution) to calculate a Z-score, which measures how many standard deviations the sample mean deviates from the population mean. The test is commonly applied in scenarios such as quality control (e.g., testing if a batch of products meets a specified standard), comparing survey results to national averages, or evaluating the effectiveness of a treatment against a known benchmark. The Z-test assumes the data is normally distributed or approximately normal due to the Central Limit Theorem for large samples. However, if the population standard deviation is unknown or the sample size is small, the t-test is more appropriate. By comparing the calculated Z-score to critical values from the Z-table (or using the p-value approach), researchers can decide whether to reject the null hypothesis, making the Z-test a powerful tool for inference in large-sample studies with known variability.

Q9.How do you calculate the Z-score, and what does it represent in hypothesis testing?
  - In hypothesis testing, the Z-score is a standardized measure that quantifies how far a sample statistic (like the mean) deviates from the hypothesized population parameter, expressed in units of standard deviation. To calculate it for a sample mean, you subtract the population mean (μ) from the observed sample mean (x̄), then divide by the standard error of the mean (σ/√n, where σ is the population standard deviation and n is the sample size). The resulting Z-score tells you how many standard deviations away from the expected value your sample result falls. In hypothesis testing, this score is compared against critical values from the standard normal distribution to determine statistical significance. For instance, a Z-score beyond ±1.96 (for α=0.05) indicates the result is statistically significant in a two-tailed test. Essentially, the Z-score bridges your raw data to probability, allowing you to assess whether observed differences are likely due to chance or reflect true population differences. It's particularly useful when working with large samples (n≥30) where the Central Limit Theorem ensures normality, and when population parameters (especially σ) are known. However, if σ is unknown or samples are small, a t-test becomes more appropriate. The Z-score's power lies in transforming diverse datasets into a common scale for probabilistic interpretation

Q10.What is the T-distribution, and when should it be used instead of the normal distribution?
  - The t-distribution (or Student's t-distribution) is a probability distribution similar to the normal (Z) distribution but with thicker tails, designed to account for greater uncertainty when working with small sample sizes or unknown population parameters. It is used when estimating the mean of a normally distributed population in scenarios where the sample size is small (typically n < 30) and the population standard deviation (σ) is unknown, requiring the sample standard deviation (s) as an estimate. Unlike the normal distribution, which assumes fixed variance, the t-distribution adjusts its shape based on degrees of freedom (df = n − 1), becoming narrower and more Z-like as sample size increases.

When to Use the t-Distribution Instead of the Normal Distribution:
Small Samples (n < 30): The t-distribution corrects for the increased variability and unreliability of small samples.

Unknown Population Standard Deviation (σ): When σ must be estimated from sample data (using *s*), the t-distribution provides more accurate confidence intervals and hypothesis tests.

Inference for Means: Commonly used in t-tests (one-sample, independent two-sample, or paired samples) and constructing confidence intervals for means.

For large samples (n ≥ 30), the t-distribution converges to the normal distribution due to the Central Limit Theorem, making the choice less critical. However, statistical software often defaults to the t-distribution for robustness, even with larger samples. Always opt for the t-distribution when dealing with small-sample inference to avoid underestimating uncertainty and reducing Type I errors.

Key Differences:

Use Cases:

It is widely applied when working with small sample sizes or when the population standard deviation (σ) is unknown. Here’s how it works and when to use it:

Q13.What is the relationship between Z-test and T-test in hypothesis testing?
  - The Z-test and t-test are closely related statistical methods used for hypothesis testing about population means, but they differ in their assumptions and applications. The Z-test relies on the normal distribution and is used when the population standard deviation (σ) is known and the sample size is large (typically n ≥ 30), leveraging the Central Limit Theorem for validity. In contrast, the t-test uses the t-distribution, which has heavier tails to account for greater uncertainty when σ is unknown and estimated from the sample, especially with small sample sizes (n < 30). As sample size increases, the t-distribution converges to the normal distribution, making the t-test results approximate those of a Z-test. Essentially, the t-test is a more adaptable version of the Z-test, designed for real-world conditions where population parameters are rarely known. Both tests follow the same procedural steps—formulating hypotheses, calculating a test statistic, and comparing it to a critical value—but the t-test’s flexibility makes it the default choice for most practical research, while the Z-test remains useful in specific, large-scale scenarios with known variance.

Q14.> What is a confidence interval, and how is it used to interpret statistical results?
 - A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain an unknown population parameter (e.g., mean, proportion) with a specified level of confidence, typically 95% or 99%. It provides a measure of uncertainty around a point estimate, offering a more informative interpretation than a single value alone. For example, a 95% CI for a mean difference of 10 units between two groups (with a CI of 7 to 13) implies that, if the study were repeated many times, 95% of such intervals would contain the true population mean difference. Unlike a p-value, which only tests statistical significance, a CI also quantifies the precision of the estimate and the practical significance of the results. If the interval excludes the null value (e.g., zero for mean differences or one for risk ratios), it indicates statistical significance at the chosen confidence level. By examining the width of the interval (narrower = more precise) and its bounds, researchers can assess both the reliability and magnitude of an effect, making CIs a cornerstone of robust statistical inference.

Q15.What is the margin of error, and how does it affect the confidence interval?
  -  The margin of error (MOE) quantifies the maximum expected difference between the observed sample statistic (e.g., mean or proportion) and the true population parameter, due to random sampling variability. It is directly tied to the confidence interval (CI), forming its bounds. For example, a 95% CI for a sample mean of 50 with an MOE of 5 is written as 50 ± 5, or [45, 55].

The MOE is influenced by three key factors:

Sample Size (n): Larger samples reduce the MOE, yielding narrower, more precise CIs (since MOE ∝ 1/√n).

Variability in Data (standard deviation, σ): Higher variability widens the MOE.

Confidence Level (e.g., 90%, 95%): Higher confidence (e.g., 99% vs. 95%) increases the MOE, producing wider intervals to capture the true parameter more reliably.

Impact on Interpretation:

A smaller MOE indicates greater precision, making the results more actionable (e.g., "The average score is 50 ± 2" is more informative than "50 ± 10").

If the MOE is too large, the CI may include null or trivial effects, undermining conclusions. For instance, a CI of [−2, 12] for a treatment effect includes zero, suggesting the effect might be negligible.

Q18.What is the Chi-square goodness of fit test, and how is it applied?
  - The Chi-square goodness of fit test is a statistical method used to determine whether observed categorical data matches an expected distribution. It compares the observed frequencies of different categories with their theoretically expected frequencies under a given hypothesis.

To apply the test, follow these steps:

Define Hypotheses –

Null Hypothesis (H₀): The observed data fits the expected distribution.

Alternative Hypothesis (H₁): The observed data does not fit the expected distribution.

Calculate Expected Frequencies – Based on the hypothesized distribution (e.g., uniform, binomial, or normal). Compare to Critical Value – Using the Chi-square distribution table with appropriate degrees of freedom (df = number of categories - 1 - number of estimated parameters).

Make a Decision – If the test statistic exceeds the critical value (or p-value < significance level, e.g., 0.05), reject H₀, indicating a poor fit.

Example: Testing if a die is fair (each face has equal probability). If observed rolls significantly deviate from expected frequencies, the die may be biased. The test helps assess such discrepancies objectively.

Q19.What is the F-distribution, and when is it used in hypothesis testing?
  - The F-distribution is a continuous probability distribution used in statistical tests involving variance comparisons, particularly in analysis of variance (ANOVA) and regression. It is characterized by two degrees of freedom—one for the numerator (between-group variability) and one for the denominator (within-group variability)—and is always right-skewed, approaching symmetry as degrees of freedom increase. The F-distribution is applied in three main scenarios: ANOVA, to test differences among group means (e.g., comparing the effectiveness of multiple treatments); variance equality tests, to determine if two populations have the same variability (e.g., assessing consistency in manufacturing processes); and regression analysis, to evaluate the overall significance of a model (e.g., checking if predictors collectively impact a dependent variable). In hypothesis testing, the F-test calculates a ratio of variances, with a high F-value (or low p-value) leading to rejection of the null hypothesis, indicating significant differences or model fit. This makes the F-distribution a fundamental tool for assessing variability and statistical relationships.

Q20.What is an ANOVA test, and what are its assumptions?
  - Analysis of Variance (ANOVA) is a statistical test used to compare the means of three or more groups to determine if at least one differs significantly. It works by partitioning total variability into between-group (due to treatment effects) and within-group (random error) variance. ANOVA is widely applied in experiments, such as comparing the effectiveness of different teaching methods or drug treatments. For valid results, ANOVA relies on three key assumptions: normality (data in each group should be normally distributed, checked via Shapiro-Wilk or Q-Q plots), homogeneity of variance (equal variances across groups, tested using Levene’s or Bartlett’s test), and independence of observations (data points must not influence each other). Violations of these assumptions may require alternative approaches, such as non-parametric tests (e.g., Kruskal-Wallis) or transformations. ANOVA is foundational in experimental research, helping researchers identify significant differences while controlling Type I error rates.

Q21.What are the different types of ANOVA tests?
  - ANOVA tests come in several types, each designed for specific experimental designs. The most common include:

One-Way ANOVA – Compares means across three or more independent groups (e.g., testing differences in exam scores among students taught by three different methods).

Two-Way ANOVA – Analyzes the effect of two independent variables and their interaction (e.g., assessing how both diet and exercise influence weight loss).

Repeated Measures ANOVA – Used for dependent samples measured multiple times (e.g., tracking patient blood pressure before, during, and after treatment).

Factorial ANOVA – Extends Two-Way ANOVA to three or more factors (e.g., studying the combined effects of temperature, humidity, and light on plant growth).

MANOVA (Multivariate ANOVA) – Tests multiple dependent variables simultaneously (e.g., comparing the impact of a teaching method on both math and science scores)

Q22.What is the F-test, and how does it relate to hypothesis testing?
  - The F-test is a statistical test that compares variances to assess significant differences between groups or models, primarily using the F-distribution. It plays a key role in hypothesis testing by evaluating whether the variability between groups (explained variance) is significantly larger than variability within groups (unexplained variance). In ANOVA, the F-test determines if group means differ (e.g., comparing treatments), while in regression, it tests the overall model significance (e.g., checking if predictors jointly affect the outcome). For equality of variances (e.g., Levene’s test), it ensures homogeneity assumptions are met. The test yields an F-statistic (ratio of variances) and a p-value; if the p-value is below the significance level (e.g., 0.05), the null hypothesis (e.g., "no difference" or "no model fit") is rejected. Thus, the F-test bridges variance analysis and hypothesis testing, enabling robust comparisons in experiments and modeling.

```python
#Q2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters for the discrete uniform distribution
low = 1      # Minimum value
high = 6     # Maximum value (inclusive)
size = 1000  # Number of random variates to generate

# Generate random variates (like rolling a fair die)
random_variates = np.random.randint(low, high + 1, size)

# Calculate PMF (equal probability for each outcome)
x = np.arange(low, high + 1)
pmf = np.ones_like(x) / len(x)  # Equal probabilities

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(x, pmf, width=0.4, color='skyblue', edgecolor='navy', alpha=0.7)
plt.title(f'PMF of Discrete Uniform Distribution ({low} to {high})', fontsize=14)
plt.xlabel('Outcomes', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.xticks(x)
plt.ylim(0, 1.1*max(pmf))
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate each bar with probability value
for xi, p in zip(x, pmf):
    plt.text(xi, p + 0.01, f'{p:.2f}', ha='center', fontsize=10)

plt.show()
```

```python
#Q3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
import numpy as np
import matplotlib.pyplot as plt

def bernoulli_pdf(p):
    """
    Calculate and plot the PDF of a Bernoulli distribution.

    Parameters:
    p (float): Probability of success (1), must be between 0 and 1

    Returns:
    dict: Dictionary containing the PDF values for k=0 and k=1
    """
    if not 0 <= p <= 1:
        raise ValueError("Probability p must be between 0 and 1")

    # Bernoulli PDF: P(k) = p^k * (1-p)^(1-k) for k ∈ {0,1}
    pdf = {0: 1 - p, 1: p}

    # Visualization
    plt.figure(figsize=(8, 5))
    plt.bar(pdf.keys(), pdf.values(), width=0.1, color=['skyblue', 'salmon'])
    plt.title(f'Bernoulli Distribution PDF (p={p})', fontsize=14)
    plt.xlabel('Outcome (k)', fontsize=12)
    plt.ylabel('Probability', fontsize=12)
    plt.xticks([0, 1])
    plt.ylim(0, 1.1)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Annotate bars
    for k, prob in pdf.items():
        plt.text(k, prob + 0.02, f'{prob:.2f}', ha='center', fontsize=12)

    return pdf

# Example usage
p = 0.3  # Probability of success
pdf_values = bernoulli_pdf(p)
plt.show()
```

```python
#Q4.> Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10       # Number of trials
p = 0.5      # Probability of success
size = 1000  # Number of simulations

# Generate binomial random variables
data = np.random.binomial(n, p, size)

# Calculate theoretical probabilities
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

# Plot histogram and theoretical PMF
plt.figure(figsize=(10, 6))

# Histogram of simulated data
plt.hist(data, bins=np.arange(-0.5, n+1.5, 1),
         density=True, alpha=0.7, color='skyblue',
         edgecolor='navy', label='Simulated Data')

# Theoretical PMF
plt.stem(x, pmf, linefmt='r-', markerfmt='ro',
         basefmt=' ', label='Theoretical PMF')

# Formatting
plt.title(f'Binomial Distribution (n={n}, p={p})', fontsize=14)
plt.xlabel('Number of Successes', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.xticks(x)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.xlim(-0.5, n+0.5)
plt.tight_layout()
plt.show()
```

```python
#Q6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters for the discrete uniform distribution
low = 1      # Minimum value
high = 6     # Maximum value (inclusive)

# Create the distribution
dist = randint(low, high + 1)

# Generate values and probabilities
x = np.arange(low - 1, high + 2)  # Include points before and after for step plot
cdf = dist.cdf(x)                  # Calculate CDF at each point

# Plotting
plt.figure(figsize=(10, 6))

# Step plot for discrete CDF
plt.step(x, cdf, where='post', color='navy', linewidth=2.5, label='CDF')

# Mark the actual points
plt.plot(x, cdf, 'o', color='red', markersize=8, label='CDF points')

# Formatting
plt.title(f'CDF of Discrete Uniform Distribution ({low} to {high})', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('F(x)', fontsize=12)
plt.xticks(np.arange(low, high + 1))
plt.yticks(np.arange(0, 1.1, 0.1))
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc='lower right', fontsize=12)

# Highlight key properties
plt.axhline(y=0, color='black', linewidth=0.5)
plt.axhline(y=1, color='black', linewidth=0.5)

# Annotate jumps
for i in range(low, high + 1):
    plt.annotate(f'F({i}) = {dist.cdf(i):.2f}',
                 xy=(i, dist.cdf(i)),
                 xytext=(5, 5),
                 textcoords='offset points',
                 bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5))

plt.tight_layout()
plt.show()
```

```python
#Q9.Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

def plot_z_scores(data, title="Z-Score Distribution"):
    """
    Calculate Z-scores and create a visualization of the standardized distribution.

    Parameters:
    data (array-like): Input dataset
    title (str): Title for the plot (default: "Z-Score Distribution")
    """
    # Calculate Z-scores
    z_scores = zscore(data)

    # Create figure with two subplots
    plt.figure(figsize=(12, 8))

    # Main plot: Histogram with KDE and Z-score indicators
    ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3, rowspan=2)

    # Histogram and KDE plot
    sns.histplot(z_scores, bins=30, kde=True,
                 color='skyblue', edgecolor='navy',
                 stat='density', label='Z-score distribution')

    # Plot standard normal distribution for comparison
    x = np.linspace(-4, 4, 1000)
    plt.plot(x, 1/np.sqrt(2*np.pi) * np.exp(-x**2/2),
             'r-', linewidth=2, label='Standard Normal')

    # Add reference lines
    for z in [-3, -2, -1, 0, 1, 2, 3]:
        linestyle = '--' if z != 0 else '-'
        alpha = 0.5 if z != 0 else 0.8
        plt.axvline(z, color='k', linestyle=linestyle,
                   alpha=alpha, linewidth=1,
                   label=f'Z = {z}' if z in [-3, 0, 3] else None)

    plt.title(title, fontsize=14)
    plt.xlabel('Z-scores', fontsize=12)
    plt.ylabel('Density', fontsize=12)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.3)

    # Boxplot to show outliers
    ax2 = plt.subplot2grid((3, 3), (2, 0), colspan=3)
    sns.boxplot(x=z_scores, color='lightgreen')
    plt.xlabel('Z-scores', fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.3)

    plt.tight_layout()
    plt.show()

    return z_scores

# Example usage
if __name__ == "__main__":
    # Generate sample data (normal distribution with outliers)
    np.random.seed(42)
    data = np.concatenate([np.random.normal(50, 10, 1000),
                          np.random.normal(150, 10, 10)])

    # Calculate and plot Z-scores
    z_scores = plot_z_scores(data, "Z-Score Analysis with Outliers")
```

```python
#Q10.Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import expon, norm

def demonstrate_clt(population_dist, sample_size=30, num_samples=1000):
    """
    Demonstrate the Central Limit Theorem using a non-normal distribution.

    Parameters:
    population_dist: The population distribution (e.g., exponential)
    sample_size: Size of each sample (default: 30)
    num_samples: Number of sample means to compute (default: 1000)
    """
    # Generate population data
    population = population_dist.rvs(size=100000)

    # Calculate sample means
    sample_means = np.array([np.mean(population_dist.rvs(size=sample_size))
                           for _ in range(num_samples)])

    # Create figure
    plt.figure(figsize=(14, 6))

    # Plot 1: Population distribution
    plt.subplot(1, 2, 1)
    sns.histplot(population, bins=50, kde=True, color='skyblue')
    plt.title(f"Population Distribution\n({population_dist.dist.name.capitalize()})", fontsize=12)
    plt.xlabel("Value")
    plt.ylabel("Density")

    # Plot 2: Distribution of sample means
    plt.subplot(1, 2, 2)
    sns.histplot(sample_means, bins=30, kde=True, color='salmon', stat='density')

    # Overlay normal distribution with CLT parameters
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    pop_mean = np.mean(population)
    pop_std = np.std(population)
    clt_std = pop_std / np.sqrt(sample_size)
    p = norm.pdf(x, pop_mean, clt_std)
    plt.plot(x, p, 'k', linewidth=2, label='CLT Prediction')

    plt.title(f"Distribution of Sample Means (n={sample_size})\nCLT Demonstration", fontsize=12)
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend()

    plt.tight_layout()
    plt.show()

    # Print CLT parameters
    print(f"Population mean: {pop_mean:.4f}")
    print(f"Population std: {pop_std:.4f}")
    print(f"CLT-predicted std of means: {clt_std:.4f}")
    print(f"Actual std of means: {np.std(sample_means):.4f}")

# Example usage with exponential distribution (λ=1)
demonstrate_clt(expon(scale=1), sample_size=30, num_samples=10000)
```

```python
#11.Simulate multiple samples from a normal distribution and verify the Central Limit TheoremE
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
population_mean = 50
population_std = 10
sample_size = 30  # Size of each sample
num_samples = 1000  # Number of samples to draw

# Generate samples
sample_means = []
for _ in range(num_samples):
    sample = np.random.normal(population_mean, population_std, sample_size)
    sample_means.append(np.mean(sample))

# Convert to numpy array
sample_means = np.array(sample_means)
plt.figure(figsize=(12, 6))

# Plot histogram of sample means
plt.subplot(1, 2, 1)
sns.histplot(sample_means, kde=True, stat='density')
plt.title(f'Distribution of Sample Means\n(n={sample_size}, {num_samples} samples)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')

# Plot Q-Q plot to check normality
plt.subplot(1, 2, 2)
stats.probplot(sample_means, dist="norm", plot=plt)
plt.title('Q-Q Plot of Sample Means')

plt.tight_layout()
plt.show()
theoretical_std = population_std / np.sqrt(sample_size)
observed_mean = np.mean(sample_means)
observed_std = np.std(sample_means, ddof=1)

# Normality test
shapiro_test = stats.shapiro(sample_means)

print(f"Population mean: {population_mean:.2f}")
print(f"Mean of sample means: {observed_mean:.2f}")
print(f"\nTheoretical standard error: {theoretical_std:.4f}")
print(f"Observed standard deviation of sample means: {observed_std:.4f}")
print(f"\nShapiro-Wilk test for normality: p-value = {shapiro_test[1]:.4f}")
```

```python
#Q13.Generate random variables and calculate their corresponding probabilities using the binomial distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def binomial_distribution_analysis(n=10, p=0.5, num_samples=1000):
    """
    Generate random binomial variables and calculate their probabilities.

    Parameters:
    n (int): Number of trials
    p (float): Probability of success in each trial (0 ≤ p ≤ 1)
    num_samples (int): Number of random variables to generate
    """
    # Generate random binomial variables
    random_vars = np.random.binomial(n, p, num_samples)

    # Calculate theoretical probabilities
    x_values = np.arange(0, n+1)
    pmf = binom.pmf(x_values, n, p)  # Probability Mass Function
    cdf = binom.cdf(x_values, n, p)  # Cumulative Distribution Function

    # Create figure
    plt.figure(figsize=(14, 5))

    # Plot 1: Histogram of random variables vs theoretical PMF
    plt.subplot(1, 2, 1)
    plt.hist(random_vars, bins=np.arange(-0.5, n+1.5, 1),
             density=True, alpha=0.7, label='Random Samples')
    plt.stem(x_values, pmf, linefmt='r-', markerfmt='ro', basefmt=' ',
             label='Theoretical PMF')
    plt.title(f'Binomial Distribution (n={n}, p={p})\nRandom Variables vs PMF')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Plot 2: Cumulative Distribution Function
    plt.subplot(1, 2, 2)
    plt.step(x_values, cdf, where='post', color='g', label='Theoretical CDF')
    plt.title(f'Binomial Distribution (n={n}, p={p})\nCumulative Distribution Function')
    plt.xlabel('Number of Successes')
    plt.ylabel('Cumulative Probability')
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Print summary statistics
    print("\n=== Binomial Distribution Summary ===")
    print(f"Parameters: n={n} trials, p={p} success probability")
    print(f"Generated {num_samples} random variables")
    print("\nTheoretical Properties:")
    print(f"Mean (μ): {n*p:.2f}")
    print(f"Variance (σ²): {n*p*(1-p):.2f}")
    print(f"Standard Deviation (σ): {np.sqrt(n*p*(1-p)):.2f}")

    # Calculate and print probabilities for each possible outcome
    print("\nProbability for each possible number of successes (k):")
    for k in x_values:
        prob = binom.pmf(k, n, p)
        print(f"P(X = {k}): {prob:.4f}")

    # Calculate probability of ranges
    print("\nProbability of ranges:")
    print(f"P(X ≤ {n//2}): {binom.cdf(n//2, n, p):.4f}")
    print(f"P(X > {n//2}): {1 - binom.cdf(n//2, n, p):.4f}")
    print(f"P({n//3} ≤ X ≤ {2*n//3}): {binom.cdf(2*n//3, n, p) - binom.cdf(n//3-1, n, p):.4f}")

# Example usage
binomial_distribution_analysis(n=10, p=0.3, num_samples=1000)
```

```python
#Q14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distributionE
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
    """Calculate the Z-score for a data point."""
    return (data_point - mean) / std_dev

def plot_comparison(z_score):
    """Plot the standard normal distribution and mark the Z-score position."""
    # Create figure
    plt.figure(figsize=(12, 6))

    # Generate x values for the plot
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)  # Standard normal PDF

    # Plot the standard normal distribution
    plt.plot(x, y, 'b-', linewidth=2, label='Standard Normal Distribution')
    plt.title('Comparison of Z-score to Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.grid(True, alpha=0.3)

    # Mark the mean (Z=0)
    plt.axvline(x=0, color='green', linestyle='--', label='Mean (μ=0)')

    # Mark the Z-score position
    plt.axvline(x=z_score, color='red', linestyle='--', label=f'Your Z-score ({z_score:.2f})')

    # Shade the relevant area
    if z_score > 0:
        x_fill = np.linspace(z_score, 4, 100)
    else:
        x_fill = np.linspace(-4, z_score, 100)
    plt.fill_between(x_fill, norm.pdf(x_fill), color='orange', alpha=0.3)

    # Calculate and display probabilities
    if z_score > 0:
        prob_left = norm.cdf(z_score)
        prob_right = 1 - prob_left
        plt.text(z_score + 0.1, 0.1, f'{prob_right:.2%} of data\nabove this point',
                ha='left', va='center', bbox=dict(facecolor='white', alpha=0.8))
    else:
        prob_left = norm.cdf(z_score)
        prob_right = 1 - prob_left
        plt.text(z_score - 0.1, 0.1, f'{prob_left:.2%} of data\nbelow this point',
                ha='right', va='center', bbox=dict(facecolor='white', alpha=0.8))

    plt.legend()
    plt.show()

def main():
    print("=== Z-Score Calculator and Standard Normal Comparison ===")

    # Get user input
    data_point = float(input("Enter the data point value: "))
    mean = float(input("Enter the mean of the distribution: "))
    std_dev = float(input("Enter the standard deviation of the distribution: "))

    # Calculate Z-score
    z_score = calculate_z_score(data_point, mean, std_dev)

    # Display results
    print(f"\nCalculated Z-score: {z_score:.4f}")

    # Interpret the Z-score
    print("\nInterpretation:")
    if abs(z_score) < 1:
        print("- Your data point is within 1 standard deviation of the mean (common occurrence)")
    elif abs(z_score) < 2:
        print("- Your data point is between 1-2 standard deviations from the mean (unusual but not rare)")
    elif abs(z_score) < 3:
        print("- Your data point is between 2-3 standard deviations from the mean (very unusual)")
    else:
        print("- Your data point is more than 3 standard deviations from the mean (extremely rare)")

    # Calculate probabilities
    prob_less = norm.cdf(z_score)
    prob_greater = 1 - prob_less

    print(f"\nProbability Analysis:")
    print(f"- Probability of value ≤ your data point: {prob_less:.4f} or {prob_less*100:.2f}%")
    print(f"- Probability of value > your data point: {prob_greater:.4f} or {prob_greater*100:.2f}%")

    # Plot comparison
    plot_comparison(z_score)

if __name__ == "__main__":
    main()
```

```python
#Q15.Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def z_test(sample, population_mean, population_std=None, sample_std=None, alpha=0.05, test_type='two-tailed'):
    """
    Perform a Z-test for hypothesis testing.

    Parameters:
    - sample: Array-like, sample data
    - population_mean: Hypothesized population mean
    - population_std: Population standard deviation (if known)
    - sample_std: Sample standard deviation (if population_std unknown)
    - alpha: Significance level (default 0.05)
    - test_type: 'two-tailed', 'right-tailed', or 'left-tailed'

    Returns:
    - z_score: Calculated Z-statistic
    - p_value: Corresponding p-value
    - conclusion: Test conclusion
    """
    sample_size = len(sample)
    sample_mean = np.mean(sample)

    # Calculate standard error
    if population_std:
        std_error = population_std / np.sqrt(sample_size)
    elif sample_std:
        std_error = sample_std / np.sqrt(sample_size)
    else:
        std_error = np.std(sample, ddof=1) / np.sqrt(sample_size)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / std_error

    # Calculate p-value based on test type
    if test_type == 'two-tailed':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    elif test_type == 'right-tailed':
        p_value = 1 - stats.norm.cdf(z_score)
    elif test_type == 'left-tailed':
        p_value = stats.norm.cdf(z_score)
    else:
        raise ValueError("test_type must be 'two-tailed', 'right-tailed', or 'left-tailed'")

    # Make conclusion
    if p_value < alpha:
        conclusion = f"Reject the null hypothesis (p = {p_value:.4f} < α = {alpha})"
    else:
        conclusion = f"Fail to reject the null hypothesis (p = {p_value:.4f} ≥ α = {alpha})"

    return z_score, p_value, conclusion

def plot_z_test(z_score, alpha=0.05, test_type='two-tailed'):
    """Visualize the Z-test results"""
    plt.figure(figsize=(10, 6))

    # Plot standard normal distribution
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)
    plt.plot(x, y, 'b-', label='Standard Normal Distribution')

    # Plot critical regions
    if test_type == 'two-tailed':
        critical_value = stats.norm.ppf(1 - alpha/2)
        left_critical = -critical_value
        plt.axvline(left_critical, color='red', linestyle='--', label='Critical Values')
        plt.axvline(critical_value, color='red', linestyle='--')
        plt.fill_between(x, y, where=(x < left_critical) | (x > critical_value),
                        color='red', alpha=0.3, label='Rejection Region')
    elif test_type == 'right-tailed':
        critical_value = stats.norm.ppf(1 - alpha)
        plt.axvline(critical_value, color='red', linestyle='--', label='Critical Value')
        plt.fill_between(x, y, where=x > critical_value,
                        color='red', alpha=0.3, label='Rejection Region')
    elif test_type == 'left-tailed':
        critical_value = stats.norm.ppf(alpha)
        plt.axvline(critical_value, color='red', linestyle='--', label='Critical Value')
        plt.fill_between(x, y, where=x < critical_value,
                        color='red', alpha=0.3, label='Rejection Region')

    # Mark the Z-score
    plt.axvline(z_score, color='green', linestyle='-', linewidth=2, label=f'Z-score ({z_score:.2f})')

    plt.title(f'Z-test Visualization ({test_type.replace("-", " ")})')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# Example usage
if __name__ == "__main__":
    print("=== Z-Test Hypothesis Testing ===")

    # Sample dataset (replace with your data)
    sample_data = np.array([28.5, 29.3, 30.1, 28.9, 29.7, 30.5, 29.2, 28.8, 30.0, 29.4])

    # Known population parameters (or None if unknown)
    pop_mean = 29.0  # Null hypothesis value
    pop_std = 0.8    # If population standard deviation is known
    sample_std = None # If population std unknown, use sample std

    # Test parameters
    significance_level = 0.05
    test_type = 'two-tailed'  # or 'right-tailed', 'left-tailed'

    print(f"\nSample mean: {np.mean(sample_data):.2f}")
    print(f"Sample size: {len(sample_data)}")
    print(f"Null hypothesis (H₀): μ = {pop_mean}")

    # Perform Z-test
    z_score, p_value, conclusion = z_test(
        sample_data, pop_mean, pop_std, sample_std, significance_level, test_type
    )

    # Display results
    print("\n=== Test Results ===")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(conclusion)

    # Visualize the test
    plot_z_test(z_score, significance_level, test_type)
```

```python
#Q16.Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_confidence_interval(data, confidence=0.95):
    """
    Calculate confidence interval for a dataset.

    Parameters:
    - data: Array-like, the sample data
    - confidence: Confidence level (default 0.95 for 95%)

    Returns:
    - (mean, margin_of_error, (lower_bound, upper_bound))
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
    sample_size = len(data)

    # Calculate standard error
    std_error = sample_std / np.sqrt(sample_size)

    # Get critical value (t-score for small samples, z-score for large samples)
    if sample_size < 30:
        critical_value = stats.t.ppf((1 + confidence) / 2, df=sample_size-1)
    else:
        critical_value = stats.norm.ppf((1 + confidence) / 2)

    # Calculate margin of error
    margin_of_error = critical_value * std_error

    # Calculate confidence interval bounds
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return sample_mean, margin_of_error, (lower_bound, upper_bound)

def plot_confidence_interval(mean, ci, true_mean=None):
    """Visualize the confidence interval"""
    plt.figure(figsize=(10, 4))

    # Plot the point estimate
    plt.plot(mean, 1, 'bo', markersize=10, label='Sample Mean')

    # Plot the confidence interval
    plt.hlines(1, ci[0], ci[1], colors='blue', linewidth=3, label='95% CI')

    # Plot true mean if provided
    if true_mean is not None:
        plt.axvline(true_mean, color='red', linestyle='--', label='True Mean')

    plt.yticks([])
    plt.xlabel('Value')
    plt.title('Confidence Interval Visualization')
    plt.legend(loc='upper right')
    plt.grid(True, alpha=0.3)
    plt.show()

# Example usage
if __name__ == "__main__":
    print("=== Confidence Interval Calculator ===")

    # Generate sample data (replace with your dataset)
    np.random.seed(42)
    true_population_mean = 50
    sample_data = np.random.normal(loc=true_population_mean, scale=10, size=50)

    # Calculate confidence interval
    confidence_level = 0.95
    mean, margin, (lower, upper) = calculate_confidence_interval(sample_data, confidence_level)

    # Display results
    print(f"\nSample size: {len(sample_data)}")
    print(f"Sample mean: {mean:.2f}")
    print(f"Sample standard deviation: {np.std(sample_data, ddof=1):.2f}")
    print(f"\n{confidence_level*100:.0f}% Confidence Interval:")
    print(f"Margin of error: ±{margin:.2f}")
    print(f"Lower bound: {lower:.2f}")
    print(f"Upper bound: {upper:.2f}")
    print(f"Range: {upper-lower:.2f}")

    # Interpretation
    print("\nInterpretation:")
    print(f"We are {confidence_level*100:.0f}% confident that the true population mean")
    print(f"lies between {lower:.2f} and {upper:.2f}.")

    # Check if true mean is captured (for demonstration)
    if 'true_population_mean' in locals():
        if lower <= true_population_mean <= upper:
            print(f"\nThe true mean ({true_population_mean}) IS within our confidence interval.")
        else:
            print(f"\nThe true mean ({true_population_mean}) IS NOT within our confidence interval.")

    # Visualize the confidence interval
    plot_confidence_interval(mean, (lower, upper), true_population_mean if 'true_population_mean' in locals() else None)
```

```python
#Q17.E Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
from scipy import stats

# Parameters
mu = 50      # True mean
sigma = 10   # True standard deviation
n = 100      # Sample size

# Generate random sample
np.random.seed(42)  # For reproducibility
sample = np.random.normal(mu, sigma, n)
x_bar = np.mean(sample)          # Sample mean
s = np.std(sample, ddof=1)       # Sample standard deviation (ddof=1 for unbiased estimate)
se = s / np.sqrt(n)              # Standard error of the mean
confidence_level = 0.95
alpha = 1 - confidence_level
dof = n - 1                      # Degrees of freedom

# Critical value (two-tailed)
t_critical = stats.t.ppf(1 - alpha/2, dof)

# Margin of error
margin_of_error = t_critical * se

# Confidence interval
ci_lower = x_bar - margin_of_error
ci_upper = x_bar + margin_of_error
print(f"Sample mean (x̄): {x_bar:.2f}")
print(f"Sample standard deviation (s): {s:.2f}")
print(f"Standard error of the mean (SE): {se:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Q18.Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0      # Mean
sigma = 1   # Standard deviation

# Generate x values (range of the distribution)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calculate the PDF (Probability Density Function)
pdf = norm.pdf(x, mu, sigma)

# Plot the PDF
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Normal Distribution (μ={mu}, σ={sigma})', color='blue', linewidth=2)

# Add title and labels
plt.title('Probability Density Function (PDF) of a Normal Distribution', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Highlight the mean and standard deviations
plt.axvline(mu, color='red', linestyle='--', label=f'Mean (μ) = {mu}')
plt.axvline(mu + sigma, color='green', linestyle=':', label=f'μ + σ = {mu + sigma}')
plt.axvline(mu - sigma, color='green', linestyle=':', label=f'μ - σ = {mu - sigma}')

# Add legend and show plot
plt.legend()
plt.show()
```

```python
#19.Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_ = 5  # Mean (λ) of the Poisson distribution
max_k = 15   # Maximum value of k (events) to consider

# Generate k values (0, 1, 2, ..., max_k)
k_values = np.arange(0, max_k + 1)

# Calculate the CDF (P(X ≤ k))
cdf = poisson.cdf(k_values, mu=lambda_)

# Plot the CDF
plt.figure(figsize=(10, 6))
plt.step(k_values, cdf, where='post', color='blue', linewidth=2, label=f'Poisson CDF (λ={lambda_})')

# Highlight key points
plt.scatter(k_values, cdf, color='red', zorder=5)  # Points at each k
plt.axhline(y=0.5, color='green', linestyle='--', label='Median (P(X ≤ k) = 0.5)')

# Add labels and title
plt.title(f'Cumulative Distribution Function (CDF) of Poisson(λ={lambda_})', fontsize=14)
plt.xlabel('Number of Events (k)', fontsize=12)
plt.ylabel('Cumulative Probability (P(X ≤ k))', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.xticks(k_values)
plt.show()

# Print CDF values
print("k\tP(X ≤ k)")
for k, prob in zip(k_values, cdf):
    print(f"{k}\t{prob:.4f}")
```

```python
#Q20.E Simulate a random variable using a continuous uniform distribution and calculate its expected value?
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the uniform distribution
a = 2.0  # Lower bound
b = 5.0  # Upper bound
sample_size = 10_000  # Number of random samples

# Generate random samples from Uniform(a, b)
samples = np.random.uniform(low=a, high=b, size=sample_size)

# Theoretical expected value (mean) of Uniform(a, b)
theoretical_mean = (a + b) / 2

# Empirical mean (average of the samples)
empirical_mean = np.mean(samples)

# Plot the histogram of samples
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black', label='Sampled Data')

# Plot the theoretical uniform PDF for comparison
x = np.linspace(a, b, 1000)
uniform_pdf = np.ones_like(x) / (b - a)  # PDF of Uniform(a, b)
plt.plot(x, uniform_pdf, 'r-', linewidth=2, label='Theoretical PDF')

# Add expected values to the plot
plt.axvline(theoretical_mean, color='green', linestyle='--', linewidth=2, label=f'Theoretical Mean = {theoretical_mean:.2f}')
plt.axvline(empirical_mean, color='black', linestyle=':', linewidth=2, label=f'Empirical Mean = {empirical_mean:.2f}')

# Add labels and legend
plt.title('Simulation of a Continuous Uniform Distribution', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Print results
print(f"Theoretical Expected Value (E[X]): {theoretical_mean:.4f}")
print(f"Empirical Mean (from simulation): {empirical_mean:.4f}")
print(f"Absolute Error: {abs(theoretical_mean - empirical_mean):.4f}")
```
